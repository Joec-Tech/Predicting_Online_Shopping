```python
def annotate_percent(ax, total_count):
    """Annotate bars with percentages."""
    for p in ax.patches:
        count = p.get_height()
        perc = 100 * count / total_count if total_count else 0
        ax.annotate(f"{perc:.1f}%",
                    (p.get_x() + p.get_width()/2, count),
                    ha="center", va="bottom")

# Step 7: % customers that bought (Revenue)
plt.figure(figsize=(8,5))
total = len(df2)
ax = sns.countplot(x="Revenue", data=df2)
annotate_percent(ax, total)
plt.title("Revenue Distribution (Buy vs Not Buy)")
plt.tight_layout()
plt.show()


```

```python
# Preparing the dataset

X = df2.iloc[:, [1, 6]].values

# Checking the shape of the dataset
X.shape

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Elbow method
wcss = []

for i in range(1, 11):
    km = KMeans(
        n_clusters=i,
        init='k-means++',
        max_iter=300,
        n_init=10,
        random_state=0,
        algorithm='elkan',
        tol=0.001
    )
    
    km.fit(X)
    wcss.append(km.inertia_)

# Plotting the elbow curve
plt.figure(figsize=(15, 7))
plt.plot(range(1, 11), wcss, marker='o')
plt.grid(True)
plt.tight_layout()
plt.title('The Elbow Method', fontsize=20)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# One-hot encoding
data1 = pd.get_dummies(df2)

# Check encoded columns
data1.columns

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# Encode target variable (Revenue) AFTER encoding predictors
data1['Revenue'] = le.fit_transform(data1['Revenue'])

# Check class balance
data1['Revenue'].value_counts()

# -------------------------------
# Getting dependent and independent variables
# -------------------------------
X = data1.drop(['Revenue'], axis=1)
y = data1['Revenue']

# -------------------------------
# Checking the shapes
# -------------------------------
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
# Splitting the data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0
)

# Checking the shapes
print("Shape of x_train:", x_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of x_test:", x_test.shape)
print("Shape of y_test:", y_test.shape)
```

```python

# RandomForest classifier model Building
# MODELLING
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
model = RandomForestClassifier()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
# evaluating the model
print("Training Accuracy :", model.score(x_train, y_train))
print("Testing Accuracy :", model.score(x_test, y_test))
```

```python
# confusion matrix: Confusion Matrix. Model accuracy is 89%.
cm = confusion_matrix(y_test, y_pred)
plt.rcParams['figure.figsize'] = (6, 6)
sns.heatmap(cm ,annot = True)
# classification report
cr = classification_report(y_test, y_pred)
print(cr)
```
